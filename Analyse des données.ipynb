```python
import yfinance as yf
```

```python
# Télécharger les données des actions de Tesla (TSLA)
tesla_data = yf.download('TSLA', start='2010-01-01', end='2023-01-01')

# Afficher les 5 premières lignes des données
print(tesla_data.head())
```

```python
import requests
```

```python
from bs4 import BeautifulSoup

# URL du site avec les données de revenus de Tesla
url = "https://finance.yahoo.com/quote/TSLA/financials"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Extraire les revenus (en fonction de la structure HTML)
revenus = soup.find_all('span', string='Total Revenue')  # Exemple, à ajuster selon le site
print(revenus)
```

```python
# Télécharger les données des actions de GameStop (GME)
gamestop_data = yf.download('GME', start='2010-01-01', end='2023-01-01')

# Afficher les 5 premières lignes des données
print(gamestop_data.head())
```

```python
# URL du site avec les données de revenus de GameStop
url_gme = "https://finance.yahoo.com/quote/GME/financials"
response_gme = requests.get(url_gme)
soup_gme = BeautifulSoup(response_gme.content, 'html.parser')

# Extraire les revenus de GameStop
revenus_gme = soup_gme.find_all('span', string='Total Revenue')  # Exemple, à ajuster
print(revenus_gme)
```

```python
import matplotlib.pyplot as plt
# Tracer les actions de Tesla
tesla_data['Close'].plot(title='Prix des actions Tesla')
plt.xlabel('Date')
plt.ylabel('Prix de clôture')
plt.show()
# Tracer les revenus de Tesla 
revenus_tesla = [31.5, 24.6, 31.5, 45.2]  # Exemple
plt.bar(['2019', '2020', '2021', '2022'], revenus_tesla, color='blue')
plt.title('Revenus de Tesla')
plt.xlabel('Année')
plt.ylabel('Revenus en milliards')
plt.show()
```

```python
# Tracer les actions de GameStop
gamestop_data['Close'].plot(title='Prix des actions GameStop')
plt.xlabel('Date')
plt.ylabel('Prix de clôture')
plt.show()
# Tracer les revenus de GameStop 
revenus_gamestop = [6.5, 5.6, 6.0, 7.2]  # Exemple
plt.bar(['2019', '2020', '2021', '2022'], revenus_gamestop, color='green')
plt.title('Revenus de GameStop')
plt.xlabel('Année')
plt.ylabel('Revenus en milliards')
plt.show()
```
